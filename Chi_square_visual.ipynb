```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the CSV file
drugs = ['ampicillin','vancomycin','streptomycin','teicoplanin','gentamicin']

for i in drugs:
  drug = i
  header_ = 0
  data = pd.read_csv('/content/drive/MyDrive/wulab_final/chi_csv/{}_chi-square.csv'.format(i),header =header_,index_col =0,encoding = 'utf-8') #header=1讓第二行作為欄位抬頭

  print("---------------{}-----------------".format(i))
  print(data)
  # Setting up the plot style
  sns.set(style="whitegrid")


  #1 Sort the data by Chi-square values and select the top 20 genes
  top_100_genes = data.nsmallest(100, 'P-value')
  top_20_genes = data.nsmallest(20,'P-value' )
  top_genes = data.sort_values('Chi2', ascending=False).head(100)
  #create csv of top100 genes
  top_100_genes.to_csv('/content/drive/My Drive/wulab_final/{}_top100genes.csv'.format(i))

  # Normalize the Chi-square values for color mapping
  norm = plt.Normalize(top_genes['Chi2'].min(), top_genes['Chi2'].max())
  colors = plt.cm.Wistia(norm(top_genes['Chi2']))

  # Create the barplot
  plt.figure(figsize=(10, 6))
  bars = sns.barplot(x='Gene', y='Chi2', data=top_genes, palette=colors)

  # Add a color bar to indicate the Chi-square values
  sm = plt.cm.ScalarMappable(cmap='Wistia', norm =norm)
  sm.set_array([])
  plt.colorbar(sm, label='Chi2')

  plt.xticks(rotation=90, fontsize=6)
  plt.title('Top 100 Genes by Chi2 Values ({})'.format(i))
  plt.xlabel('Gene')
  plt.ylabel('Chi2 Values')
  plt.tight_layout()
  plt.show()


  # Sort the data by P-value to get the top 100 important genes (smallest P-values)

  print('Top 20 genes:\n', top_20_genes)
  plt.figure(figsize=(10, 6))
  sns.scatterplot(x='Gene', y='P-value', data=top_100_genes, hue='P-value', palette='deep',cmap='viridis', legend=False)
  #plt.scatter(top_100_genes['Gene'], top_100_genes['P-value'], marker ='o', cmap='viridis', color='blue', s=10)
  plt.yscale('log')  # Using a log scale for better visualization of P-values
  plt.xlabel('Gene')
  plt.ylabel('P-value')
  plt.title('P-value Distribution for Top 100 Important Genes ({})'.format(i))
  plt.xticks(rotation=90, fontsize=6)  # Rotate x-axis labels for better readability
  plt.grid(True)
  plt.tight_layout()


  # 2. Scatter plot of Chi-square values versus P-values
  plt.figure(figsize=(10, 6))
  sns.scatterplot(x='Chi2', y='P-value', data=data)
  plt.title('Chi-square Values vs. P-values ({})'.format(i))
  plt.xlabel('Chi-square Value')
  plt.ylabel('P-value')
  plt.yscale('log')  # Log scale for P-values for better visualization
  plt.tight_layout()
  plt.show()

  # 3. Highlight significant genes (P-value < 0.05)
  significant_genes = data[data['P-value'] < 0.05]

  plt.figure(figsize=(10, 6))
  sns.scatterplot(x='Chi2', y='P-value', data=significant_genes, hue='Gene', palette='deep', legend=False)
  plt.title('Significant Genes (P-value < 0.05) ({})'.format(i))
  plt.xlabel('Chi-square Value')
  plt.ylabel('P-value')
  plt.yscale('log')  # Log scale for P-values for better visualization
  plt.tight_layout()
  plt.show()
```
